```python
##Analysis by Student A
import pandas as pd
import numpy as np

# Load the combined data from the CSV file
columns_to_read = [0, 3, 4, 5, 6]
df = pd.read_csv(r"C:\Users\Aishwarya_94\Jupyter_python\final_data_combined.csv", usecols=columns_to_read)
display(df)
import pandas as pd
```

```python
import matplotlib.pyplot as plt

xi = 1
yi = 2
x = df.iloc[:, xi]
y = df.iloc[:, yi]

xl = df.columns[xi]
yl = df.columns[yi]

p = np.polyfit(x, y, 1)
fit_line = np.polyval(p, x)

# Plot the scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(x, y, color='blue', alpha=0.5)
plt.plot(x, fit_line, color='red', label='1st Order Fit')
plt.xlabel(xl)
plt.ylabel(yl)
# plt.title("Scatter Plot of 2nd vs 5th Column")
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.hist(df.iloc[:, 4], bins=5, color='blue', alpha=0.7)
plt.xlabel(df.columns[4])
plt.ylabel('Frequency')
plt.title('Histogram of share of Renewable energy (%)')
plt.grid(True)
```

```python
# Sort the DataFrame by greenhouse emissions (2nd column)
sorted_df = df.sort_values(by=df.columns[1], ascending=False)

# Select the top 5 countries
top_5_df = sorted_df.head(5)

# Extract country names and greenhouse emissions values
countries = top_5_df.iloc[:, 0]
emissions = top_5_df.iloc[:, 1]

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(countries, emissions, color='skyblue')
plt.xlabel('Countries')
plt.ylabel('Greenhouse Gas Emissions')
plt.title('Top 5 Countries Emitting Greenhouse Gases')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```

```python
import matplotlib.pyplot as plt

# Filter DataFrame to exclude rows where emissions are less than 100
filtered_df = df[df.iloc[:, 1] >= 100]

# Sort the filtered DataFrame by greenhouse emissions (2nd column) in ascending order
sorted_df = filtered_df.sort_values(by=df.columns[1])

# Select the bottom 5 countries
bottom_5_df = sorted_df.head(5)

# Reverse the order of the DataFrame
bottom_5_df = bottom_5_df[::-1]

# Extract country names and greenhouse emissions values
countries = bottom_5_df.iloc[:, 0]
emissions = bottom_5_df.iloc[:, 1]

# Plot bar chart
plt.figure(figsize=(10, 6))
plt.bar(countries, emissions, color='lightgreen')
plt.xlabel('Countries')
plt.ylabel('Greenhouse Gas Emissions')
plt.title('Bottom 5 Countries Emitting Greenhouse Gases (Emissions >= 100 Tonnes)')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping of labels
plt.show()
```
