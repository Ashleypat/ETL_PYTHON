```python


fig_xpaths = [xpath_fig_renewable, xpath_fig_hydro_g, xpath_fig_hydro_gp, xpath_fig_hydro_c, xpath_fig_wind_g, xpath_fig_wind_gp, xpath_fig_wind_c, xpath_fig_solar_g, xpath_fig_solar_gp, xpath_fig_solar_c]
button_xpaths = [xpath_button_renewable, xpath_button_hydro_g, xpath_button_hydro_gp, xpath_button_hydro_c, xpath_button_wind_g, xpath_button_wind_gp, xpath_button_wind_c, xpath_button_solar_g, xpath_button_solar_gp, xpath_button_solar_c]
table_xpaths = [xpath_table_renewable, xpath_table_hydro_g, xpath_table_hydro_gp, xpath_table_hydro_c, xpath_table_wind_g, xpath_table_wind_gp, xpath_table_wind_c, xpath_table_solar_g, xpath_table_solar_gp, xpath_table_solar_c]

# Initialize empty list to store DataFrames
dfs = []
selected_columns = ['Country/area', '2022']  

for fig_xpath, button_xpath, table_xpath in zip(fig_xpaths, button_xpaths, table_xpaths):
    
    figure = driver.find_element(By.XPATH, fig_xpath)
    
    # Scroll to the figure
    driver.execute_script("arguments[0].scrollIntoView();", figure)
    
    time.sleep(sleep_time)
    
    button = driver.find_element(By.XPATH, button_xpath)
    button.click()
    time.sleep(sleep_time)
    
    # Wait for the table to be visible
    table = WebDriverWait(driver, sleep_time).until(EC.visibility_of_element_located((By.XPATH, table_xpath)))
    
    # Extract column headers
    headers = ['Country/area', '2022']  
    
    # Extract table data into a list of lists
    data = []
    for row in table.find_elements(By.TAG_NAME, "tr")[1:]:
        row_data = [cell.text for cell in row.find_elements(By.TAG_NAME, "td")]
        selected_data = [row_data[0], row_data[1]]  # Selecting the first two columns
        data.append(selected_data)
    
    # Create a pandas DataFrame
    df = pd.DataFrame(data, columns=headers)
    
    # Append DataFrame to the list
    dfs.append(df)
```

```python
# Dirtying the data by infusing  impurities
import numpy as np

#1. Strip percentage symbol and convert to numeric 
def convert_to_numeric(df):
    for col in df.columns[1:]:
        # df[col] = pd.to_numeric(df[col].astype(str).str.rstrip('%'), errors='coerce')
        df[col] = df[col].str.replace('TWh', '').str.replace('%', '').str.replace(',', '')
        df[col] = df[col].astype('float')
    return df

# Function to introduce impurities
def introduce_impurities(df):
    
    # Generate random indices
    num_impurities = 5  # Number of locations to introduce impurities
    num_rows, num_cols = df.shape
    impurity_indices = [(np.random.randint(num_rows), np.random.randint(1, num_cols)) for _ in range(num_impurities)]
    
    # 2.Introduce impurities at random indices
    for row_idx, col_idx in impurity_indices:
        df.iloc[row_idx, 1] += 100 # Add 100 to existing value (%)

    # 3.Adding Outliers
    df.iloc[0, 2] = -10  # Adding negetaive values to %hydrogen generation
    
    # 4.Add unnecessary extra column
    df = add_extra_column(df)
    
    # 5.Duplicate the second row
    df = duplicate_second_row(df)

    # 6.Introduce Spelling mistakes to country names
    df = country_name_spelling_error(df)

    return df

#  Function to duplicate the first two rows in the DataFrame
def duplicate_second_row(df):
    second_row = df.iloc[1:2] # Extract the second row
    
    # Concatenate the second row after the first row
    df = pd.concat([df.iloc[:1], second_row, df.iloc[1:]])
    
    # Reset index to avoid index duplication
    df.reset_index(drop=True, inplace=True)
    return df

# Make spelling mistakes in the names of some countries
def country_name_spelling_error(df):
    replace_dict = {
        'Czechia': 'Shechia', 
        'Ecuador': 'Equador', 
        'Luxembourg': 'Luxemberg', 
        'Morocco': 'Moroco'
    }
    df['Country/area'] = df['Country/area'].replace(replace_dict)
    return df
    
#  Add an unnecessary column with random values
def add_extra_column(df):
    df['Unnecessary_Column'] = np.random.uniform(0, 100, size=len(df))
    return df
```

```python
# Introduce impurities into the DataFrame
import pandas as pd
import numpy as np

# Load the raw data from the CSV file
df_with_impurities = pd.read_csv("RenewEnergy_merged.csv")

pd.set_option('display.max_rows', 20)
df_with_impurities = convert_to_numeric(df_with_impurities)
df_with_impurities = introduce_impurities(df_with_impurities)
print("DataFrame with impurities:")
display(df_with_impurities)
```

```python
# Cleaning the data
import pandas as pd
import numpy as np

def clean_data(df):

    for col in df.columns[1:]:
        df[col] = pd.to_numeric(df[col].astype(str).str.rstrip('%'), errors='coerce')
        
    # 1. Replace impurities with NaN if values are above 100 in selected columns
    selected_columns = [1, 3, 4, 6, 7, 9, 10]
    df.iloc[:, selected_columns] = df.iloc[:, selected_columns].where(df.iloc[:, selected_columns] <= 100, np.nan)

    # 2. Replace impurities with NaN if values are less than 0
    df.iloc[:, 1:] = df.iloc[:, 1:].mask(df.iloc[:, 1:] < 0)
    
    # 3. Clean Duplicated Data
    df = df.drop_duplicates()  # Remove duplicate rows

    # 4. Fix spelling mistakes in country names
    replace_dict = {
        'Shechia': 'Czechia', 
        'Equador': 'Ecuador', 
        'Luxemberg': 'Luxembourg', 
        'Moroco': 'Morocco'
        }
    df.loc['Country/area'] = df['Country/area'].replace(replace_dict)
    
    # 5. Remove the unnecessary column
    df.drop(columns='Unnecessary_Column', inplace=True)

    return df
```

```python
# Clean the data
import pandas as pd
import numpy as np

df_with_impurities = pd.read_csv("RenewEnergy_merged_impure.csv")

df_cleaned = clean_data(df_with_impurities)

# Display the cleaned DataFrame
print("DataFrame after cleaning:")
display(df_cleaned)
```
